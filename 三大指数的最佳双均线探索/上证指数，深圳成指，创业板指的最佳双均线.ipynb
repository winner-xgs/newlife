```python
import pandas as pd
import numpy as np
from pylab import plt, mpl
plt.style.use('seaborn')
import akshare as ak
```

```python
from itertools import product
```

```python
data=ak.stock_zh_index_daily_em(symbol="sh000001")
data.to_excel('上证指数数据.xlsx')
```

```python
data=pd.read_excel('上证指数数据.xlsx')
data
```

```python
sma1=range(3,30,1)
sma2=range(60,550,10)
```

```python
results=pd.DataFrame()
for SMA1,SMA2 in product(sma1,sma2):
    data3=pd.DataFrame(data['close'])
    data3.dropna(inplace=True)
    data3['returns']=np.log(data3['close']/data3['close'].shift(1))
    data3['SMA1']=data3['close'].rolling(SMA1).mean()
    data3['SMA2']=data3['close'].rolling(SMA2).mean()
    data3.dropna(inplace=True)
    data3['position']=np.where(data3['SMA1']>data3['SMA2'],1,0)
    data3['strategy']=data3['position'].shift(1)*data3['returns']
    data3.dropna(inplace=True)
    pref=np.exp(data3[['returns','strategy']].sum())
    results=results.append(pd.DataFrame(
    {
        'SMA1':SMA1,'SMA2':SMA2,
        'MARKET':pref['returns']
        ,'strategy':pref['strategy'],
        'out':pref['strategy']-pref['returns']
    },index=[0]),ignore_index=True)
results.sort_values('out',ascending=False).head(8)
```
