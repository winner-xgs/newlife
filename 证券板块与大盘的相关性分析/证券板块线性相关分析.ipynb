```python
data=pd.read_excel('证券板块.xlsx')
data
```

```python
data_sz=pd.read_excel('上证指数数据.xlsx')
data_sz
```

```python
data.columns=['index','date','open','high','low','close','vol','volm']
data['date']=data['date'].apply(lambda x: x.strftime('%Y-%m-%d'))
data
```

```python
data.set_index('date',inplace=True)
data
```

```python
data_qs=pd.DataFrame(data['close'])
data_qs
```

```python
data_sz.set_index('date',inplace=True)
data_sz=pd.DataFrame(data_sz['close'])
data_sz
```

```python
data_sz.columns=['szzs']
data_qs.columns=['qs']
data_sz=data_sz['2007-08-01':]
data_sz
```

```python
data_qs=data_qs.fillna(method='ffill')
data_sz=data_sz.fillna(method='ffill')
# data_qs['2008-11-01':]
```

```python
data_f=pd.concat([data_qs,data_sz],axis=1)
data_f.sort_index(inplace=True,axis=0)
data_f
```

```python
data_f=data_f.fillna(method='ffill')
data_f
```

```python
data_f.plot(subplots=True,figsize=(10,6))
```

```python
data_f.plot(secondary_y='szzs',figsize=(10,6))
```

```python
rets=np.log(data_f/data_f.shift(1))
rets
```

```python
pd.plotting.scatter_matrix(rets,alpha=0.2,diagonal='hist',hist_kwds={'bins':35},figsize=(20,16))
```

```python
pd.plotting.scatter_matrix(rets,alpha=0.2,diagonal='kde',hist_kwds={'bins':35},figsize=(20,16))
```

# OLS Regression

```python
#最小平方法 numpy来做配适  deg是1 就是最小平方法 最后得到一次的函数进行适配
rets.dropna(inplace=True)
```

```python
reg=np.polyfit(rets['qs'],rets['szzs'],deg=1)
reg
```

```python
ax=rets.plot(kind='scatter',x='qs',y='szzs',figsize=(10,6))
ax.plot(rets['qs'],np.polyval(reg,rets['qs']),'r',lw=2)
```

```python
rets.corr()
```
